```python
import gradio as gr
```

```python
def heart(Age,Height,Weight,FCVC,NCP,CH2O,FAF,TUE,Gender,CALC,FAVC,SCC,SMOKE,family_history_with_overweight,CAEC,MTRANS):
    Ans =LR.predict(np.array([Age,Height,Weight,FCVC,NCP,CH2O,FAF,TUE,Gender,CALC,FAVC,SCC,SMOKE,family_history_with_overweight,CAEC,MTRANS]).reshape(1, -1))[0]
    if(Ans == 0):
      return 'Insufficient_Weight'
    elif(Ans == 1):
      return 'Normal_Weight'
    elif(Ans == 2):
      return 'Obesity_Type_I'
    elif(Ans == 3):
     return 'Obesity_Type_II'
    elif(Ans == 4):
      return 'Obesity_Type_III'
    elif(Ans == 5):
      return 'Overweight_Level_I'
    elif(Ans == 6):
      return 'Overweight_Level_II'
    else:
      return 'Can`t Predict'
```

```python
iface = gr.Interface(
    fn=heart,
    inputs=["number", "number","number", "number","number", "number","number", "number","number","number","number","number","number","number", "number","number"],
    outputs="text",
    title="Obesity Prediction",
)
```
